# MyLake - PySpark Notebook

Este notebook está configurado para trabajar con PySpark y los datos de MyLake.

```python
# Crear un DataFrame de ejemplo
data = [
    ("Alice", 25, "Engineering"),
    ("Bob", 30, "Sales"),
    ("Charlie", 35, "Marketing"),
    ("Diana", 28, "Engineering"),
]

columns = ["name", "age", "department"]

df = spark.createDataFrame(data, columns)
df.show()
```

```python
# Operaciones con DataFrame
df.groupBy("department").count().show()
```

```python
# SQL con Spark
df.createOrReplaceTempView("employees")

spark.sql("SELECT * FROM employees WHERE age > 25").show()
```

```python
# Guardar datos en Parquet (formato de Lakehouse)
df.write.mode("overwrite").parquet("/home/jovyan/work/employees.parquet")
print("Datos guardados en formato Parquet")
```

```python
# Leer desde Parquet
df_parquet = spark.read.parquet("/home/jovyan/work/employees.parquet")
df_parquet.show()
```
